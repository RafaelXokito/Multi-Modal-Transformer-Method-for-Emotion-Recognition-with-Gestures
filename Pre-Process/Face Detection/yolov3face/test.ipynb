```python
# import libraries
from yoloface import face_analysis
import numpy
import cv2

face=face_analysis()        #  Auto Download a large weight files from Google Drive.
                            #  only first time.
                            #  Automatically  create folder .yoloface on cwd.
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
image_path = '22_Picnic_Picnic_22_10.jpg'
```

```python
img,box,conf=face.face_detection(image_path=image_path,model='full')
print(box)                  # box[i]=[x,y,w,h]
print(conf)                 #  value between(0 - 1)  or probability
output_frame = face.show_output(img,box,frame_status=True)
```

```python
OriginalImg = cv2.imread(image_path)

i = 0

x = box[i][0]
y = box[i][1]
w = box[i][2]
h = box[i][3]

cropped_image = OriginalImg[x:x+w, y:y+h]
print([x,y,w,h])
#plt.axis("off") # remove numbered axes
#plt.imshow(cropped_image) # BGR
plt.imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB)) # RGB
cv2.imwrite('contour1.png', cropped_image)
```

```python
def cut_image_by_bbox(image, bbox):
    x, y, w, h = bbox
    return image[y:y+h, x:x+w]
```

```python
OriginalImg = cv2.imread(image_path)
cropped_image = cut_image_by_bbox(OriginalImg, box[0])
```
